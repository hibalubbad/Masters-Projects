```python
f, ax = plt.subplots(2, 3, figsize=(30, 15))
f.delaxes(ax[1,2])
ax = ax.ravel()
j = 0
column_list = ['Two Wheeled Motor Vehicles', 'Cars and Taxis', 'Buses and Coaches', 'LGV\'s', 'All HGV\'s']

for col in df_multi.iloc[: ,1:6]:
    
    new_model = Prophet(interval_width=0.95,n_changepoints=15, yearly_seasonality=True, weekly_seasonality=False, daily_seasonality=False)
    
    df_multi[['y', 'ds']] = df_multi[[col, 'year']]
    new_model.fit(df_multi)
    future = new_model.make_future_dataframe(periods=30, freq='Y')
    forecast = new_model.predict(future)

    fig = new_model.plot(forecast[19:], ax = ax[j])
    ax[j].set_title(column_list[j] + ' Miles Forecast')
    ax[j].legend(('Actual Values', 'Forecast', 'Confidence Interval'))
    j += 1

plt.show()
```

```python
import seaborn as sns
df_new['total_motor_miles'] = df_new['all_motor_vehicles_miles'] * 365
plt.figure(figsize=(18,12))
ax = sns.barplot(x='region_id', y='total_motor_miles', data=df_new)

plt.xlabel('Region')

plt.ylabel('Vehicle miles', fontsize=15)
plt.title('Average milage per region for all motor vehicle types', fontsize=20)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.show()
```

```python
avg_millage_region_all = df_new.groupby(['region_id']).mean()
## Calculating rations
avg_millage_region_all['millage_ratio'] = (avg_millage_region_all['all_motor_vehicles_miles'] 
                                              / avg_millage_region_all['all_motor_vehicles_miles'].sum())

## Additional generation required capacity at year 2040 is 103.03892831
avg_millage_region_all['additional_power_2040'] = avg_millage_region_all['millage_ratio'] * 36 * 1000000
```

```python
plt.figure(figsize=(18,12))
ax = sns.barplot(x=avg_millage_region_all.index, y='additional_power_2040', data=avg_millage_region_all)

plt.xlabel('Region', fontsize=15)
plt.ylabel('MWh', fontsize=15)
plt.title('Additional generation needed per region for year 2040', fontsize=20)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.show()
```

```python
avg_millage_region_all['solar_power_cost_only'] = avg_millage_region_all['additional_power_2040'] * 33 / 1000000

plt.figure(figsize=(18,12))
ax = sns.barplot(x=avg_millage_region_all.index, y='solar_power_cost_only', data=avg_millage_region_all)

plt.xlabel('Region', fontsize=15)
plt.ylabel('Â£ (millions)', fontsize=15)
plt.title('Large-Scale solar costs per region', fontsize=20)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.show()
```
