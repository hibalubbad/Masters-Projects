```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df =  pd.read_csv("dft_traffic_counts_aadf.csv")
```

Drop insignificant columns for analysis 

```python
df.drop('id', axis = 1, inplace = True)
df.drop('ramp', axis = 1, inplace = True)
df.drop(["hgvs_2_rigid_axle","hgvs_3_rigid_axle","hgvs_4_or_more_rigid_axle","hgvs_5_articulated_axle","hgvs_6_articulated_axle","hgvs_3_or_4_articulated_axle"], axis = 1, inplace = True)
```

```python
df = df[df['link_length_miles'].notna()]

df_grouped = df.groupby(['year', 'region_id'])
df_grouped = df_grouped.apply(lambda x: x) 
df_grouped 

type_list = ['pedal_cycles', 'two_wheeled_motor_vehicles', 'cars_and_taxis', 'buses_and_coaches', 'lgvs', 'all_hgvs', 'all_motor_vehicles']

for types in type_list:
    df_grouped[types + '_miles'] = (df_grouped[types] * df_grouped['link_length_miles'])

df_grouped = df_grouped.groupby(['year', 'region_id']).sum()

type_miles_list = ['pedal_cycles_miles', 'two_wheeled_motor_vehicles_miles', 'cars_and_taxis_miles', 'buses_and_coaches_miles', 'lgvs_miles', 'all_hgvs_miles', 'all_motor_vehicles_miles']

df_grouped[[*type_miles_list]]
```

```python
df_grouped
```

```python
df_plot = df_grouped[[*type_miles_list]].copy()
df_plot['total_motor_miles'] = df_plot['all_motor_vehicles_miles'] * 365
df_plot = df_plot.unstack(level=1)
df_plot.plot(y='total_motor_miles', figsize=(20,10))

#organizing the plot 
plt.gca().set_title('Total Number of Miles Across All Regions') 
plt.gca().set_xticks(df_plot.index.values) 
plt.gca().set_xlabel("Year") 
plt.gca().set_ylabel("Number of Miles (x$10^{10}$)")  
plt.legend(bbox_to_anchor=(1, 1))  
plt.show() 
```
